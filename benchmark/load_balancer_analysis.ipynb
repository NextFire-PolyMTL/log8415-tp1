```python
import numpy as np
import json
import matplotlib.pyplot as pyplot
```

```python
with open('../metrics/cloudwatch/load_balancer/active_connection_count_metric.json', 'r') as f:
      data = json.load(f)
```

```python
active_connection_timestamps = list(reversed(['2023/09/30 18:34', '2023/09/30 18:32', '2023/09/30 18:07', '2023/09/30 18:00', '2023/09/30 17:54', '2023/09/30 17:53', '2023/09/30 17:52', '2023/09/30 17:51', '2023/09/30 17:50', '2023/09/30 17:49', '2023/09/30 17:48', '2023/09/30 17:47'])) 
active_connection_values = list(reversed([3.0, 4.0, 2.0, 2.4, 7.8, 3.0, 6.1, 8.0, 13.0, 2.0, 18.0, 20.0])) 
x_axis = np.arange(12)
```

```python
# function to add value labels
def addMyLabels(x,y):
    for i in range(len(x)):
        pyplot.text(i, y[i], y[i], ha = 'center')
```

```python
fig = pyplot.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.bar(active_connection_timestamps, active_connection_values)
ax.set_xticklabels(active_connection_timestamps,rotation = 45)
addMyLabels(active_connection_timestamps, active_connection_values)
pyplot.xlabel("Timestamps")
pyplot.ylabel("# of active TCP Connection per times")
pyplot.plot(x_axis,active_connection_values, color="green")
pyplot.savefig('../docs/figures/active_connection_count.pdf')
pyplot.show()
```
